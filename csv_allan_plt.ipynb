```python
#load data from csv file and convert to numpy arrays
import pandas as pd
import numpy as np
import allantools
import matplotlib.pyplot as plt

file_path = "locked_freq.csv"

df = pd.read_csv(file_path)

time_elapsed = df["elapsed"].to_numpy()
delta_t = df["delta_s"].to_numpy()
frequency = df["frequency_hz"].to_numpy()
```

```python
tt = time_elapsed[int(93934-(20*60/0.5)):int(93935)]
data = frequency[int(93934-(20*60/0.5)):int(93935)]
plt.plot(tt, data)
```

```python

freq_mean = np.mean(frequency)
frac_freq = frequency / freq_mean - 1.0
sample_rate = 1.0/ np.mean(delta_t)

taus, adev, adev_err, n = allantools.adev(
    frac_freq,
    rate=sample_rate,
    data_type="freq"
)

plt.figure()
plt.loglog(taus, adev, marker='o', linestyle='-')
plt.xlabel('Tau (s)')
plt.ylabel('Allan deviation')
plt.title('Allan Deviation of Frequency Data')
plt.grid(True, which='both', linestyle='--', alpha=0.5)
plt.show()
```

```python
sample_rate = 1.0 / np.mean(delta_t)

taus, adev, adev_err, n = allantools.oadev(
    frequency,
    rate=sample_rate,
    data_type='freq'
)

plt.figure()
plt.loglog(taus, adev, marker='o', linestyle='-')
plt.xlabel('Tau (s)')
plt.ylabel('Allan deviation')
plt.title('Allan Deviation of Frequency Data')
plt.grid(True, which='both', linestyle='--', alpha=0.5)
plt.show()
```

```python
sample_rate = 1.0 / np.mean(delta_t)

taus, adev, adev_err, n = allantools.oadev(
    frac_freq,
    rate=sample_rate,
    data_type='freq'
)

plt.figure()
plt.loglog(taus, adev, marker='o', linestyle='-')
plt.xlabel('Tau (s)')
plt.ylabel('Allan deviation')
plt.title('Allan Deviation of Frequency Data')
plt.grid(True, which='both', linestyle='--', alpha=0.5)
plt.show()
```

```python
taus, totdev, totdev_err, n = allantools.totdev(
    frequency,
    rate=sample_rate,
    data_type="freq"
)

plt.figure()
plt.loglog(taus, totdev, marker='o', linestyle='-')
plt.xlabel('Tau (s)')
plt.ylabel('Total Allan deviation')
plt.title('Total Allan Deviation of Frequency Data')
plt.grid(True, which='both', linestyle='--', alpha=0.5)
plt.show()
```

```python
taus, mdev, mdev_err, n = allantools.mdev(
    frequency,
    rate=sample_rate,
    data_type="freq"
)

plt.figure()
plt.loglog(taus, mdev, marker='o', linestyle='-')
plt.xlabel('Tau (s)')
plt.ylabel('Modified Allan deviation')
plt.title('Modified Allan Deviation of Frequency Data')
plt.grid(True, which='both', linestyle='--', alpha=0.5)
plt.show()
```
